# Setting limits to the data

```python
# capping data, i.e reducing data to the 1st and 99th percentile in order to avoid extreme values. 
# (fixing extreme values to make sure they arenâ€™t too different from the rest)

def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower_bound, upper_bound)

columns_to_cap = ['trestbps', 'chol', 'oldpeak', 'thalach']
for col in columns_to_cap:
    cap_outliers(df, col)
```

# Rescaling The Data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE  # to use SMOTE 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# getting the numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the numerical columns
df_scaled = df.copy()
df_scaled[numerical_cols] = scaler.fit_transform(df[numerical_cols])  
# rescaling the data, to change the range of extreme values to fit within a certain range 
# this ensures that all the data in the data frame contributes to the machine learning process

print("\nScaled Data:")
print(df_scaled)
```

# Stacking Ensemble

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

estimators = [
    ('knn', KNeighborsClassifier(n_neighbors=7)), 
    ('dt', DecisionTreeClassifier(random_state=42))
]

# Defining the Stacking Classifier
stacking_model = StackingClassifier (
    estimators = estimators,
    final_estimator=LogisticRegression()
)

# Fit the model
stacking_model.fit(x_train, y_train)

# Predictions for both training and test sets
y_train_pred = stacking_model.predict(x_train)
y_test_pred = stacking_model.predict(x_test)

# Calculate accuracy for both sets
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Stacking - Training Accuracy: {train_accuracy:.2f}")
print(f"Stacking - Test Accuracy: {test_accuracy:.2f}")
```

```python
# Define the bins and labels
bins = [0, 40, 60, 100]  # Age ranges: 0-40, 40-60, 60-100
labels = ['young', 'middle-aged', 'senior']  # Labels for the bins

# Create a new column with age categories
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)


age_target_counts = df.groupby(['age_group', 'target']).size().unstack().fillna(0)

# Plotting
age_target_counts.plot(kind='bar', stacked=True)
plt.title('Count of Target by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Target', loc='upper right')

# Show theÂ plot
plt.show()
```

```python
# Analyze the percentage of heart disease cases across different age groups.

# Define age groups (e.g., <40, 40-50, 50-60, 60+)
bins = [20, 40, 50, 60, 70, 80]
labels = ['<40', '40-50', '50-60', '60-70', '70+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Percentage of heart disease in each age group
age_heart_disease = df.groupby('age_group')['target'].mean() * 100
age_heart_disease.plot(kind='bar', color='lightgreen')
plt.title('Heart Disease by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage with Heart Disease')
plt.show()
```

```python
# Analyze how gender and age group interact to affect heart disease risk.

# Pivot table to see the interaction
gender_age_heart_disease = df.pivot_table(values='target', index='age_group', columns='sex', aggfunc='mean') * 100
gender_age_heart_disease.plot(kind='bar', figsize=(10, 6), color=['pink', 'blue'])
plt.title('Heart Disease by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Percentage with Heart Disease')
plt.show()
```
